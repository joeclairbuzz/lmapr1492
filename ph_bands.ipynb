```python
from pymatgen.ext.matproj import MPRester
from pymatgen.phonon.plotter import PhononBSPlotter
import numpy as np 
import matplotlib.pyplot as plt
from pymatgen.core.lattice import Lattice
```

```python
mp_key = "8ChDIqwByRyXhuV8q0dfcUVnIxLTB4o7"
mp_id = "mp-961649"
```

```python
with MPRester(mp_key) as m: 
    structure = m.get_phonon_bandstructure_by_material_id(mp_id) #téléchargement de la structure
```

## Représentation de la structure de bandes de phonons du matériau

```python
bplot = PhononBSPlotter(structure)
bplot.get_plot()
```

## Détermination de la vitesse du son 

### 1 ere branche acoustique (Γ-X )

```python
data =bplot.bs_plot_data()
distances = data.get("distances")
frequency = data.get("frequency")

bplot.get_plot()

d1 = distances[0]
f1 = frequency[0][1]
x1 = distances[0][4] - distances[0][0]
y1 = f1[4] - f1[0]
v1 = 2*np.pi * abs(y1)*10**12*10**(-10)/x1 #changement d'unité et calcul de la vitesse du son car la fréquence est en terrahertz 10**12 et vecteur d'onde en angstrom 10**-10
print("La vitesse du son de la branche acoustique numéro 1 dans la direction Γ - X vaut ", v1, " m/s et est représentée par la flèche rouge.")

plt.arrow(d1[0], f1[0], x1, y1, color = 'red', width = 0.03, length_includes_head = True)
plt.show()
```

### 2 ème branche acoustique (K-Γ)

```python

bplot.get_plot()

d2 = distances[3]
f2 = frequency[3][0]
x2 = d2[-1] - d2[-5]
y2 = f2[-1] - f2[-5]
v2 = 2*np.pi * abs(y2)*10**12*10**(-10)/x2 #changement d'unité et calcul de la vitesse du son
print("La vitesse du son de la branche acoustique numéro 0 dans la direction K - Γ vaut ", v2, " m/s et est représentée par la flèche verte.")

plt.arrow(d2[-5], f2[-5], x2, y2, color = 'green', width = 0.03, length_includes_head = True)
plt.show()
```

### 3ème branche acoustique (Γ-L)

```python
bplot.get_plot()

d3 = distances[4]
f3 = frequency[4][2]
x3 = d3[4] - d3[0]
y3 = f3[4] - f3[0]
v3 = 2*np.pi * abs(y3)*10**12*10**(-10)/x3 #changement d'unité et calcul de la vitesse du son
print("La vitesse du son de la branche acoustique numéro 2 dans la direction Γ - L vaut ", v3, " m/s et est représentée par la flèche orange.")

plt.arrow(d3[0], f3[0], x3, y3, color = 'orange', width = 0.03, length_includes_head = True)
plt.show()
```
